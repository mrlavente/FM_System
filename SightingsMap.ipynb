```python
import sqlite3
import pandas as pd

connection = sqlite3.connect("database.db")
cursor = connection.cursor()
query = "SELECT * FROM Data"
cursor.execute(query)
result = cursor.fetchall()
for item in result:
    print(item)
df = pd.read_sql_query(query,connection)
df
df.to_csv('data.csv')


cursor.close()
```

```python
import folium
import pandas as pd
import numpy as np

data = pd.read_csv('data.csv')
PANAY =[11.102947,122.516279]
Map=folium.Map(PANAY, zoom_start=9, tiles='Stamen Terrain')
Map

for i in range(0,len(data)):

    folium.Marker([data.iloc[i]["coordinates_lat"],data.iloc[i]["coordinates_lng"]],tooltip=data.iloc[i]["species_name"],popup=data.iloc[i]["no_observation"]).add_to(Map)
Map
```

```python
import folium
import pandas as pd
import numpy as np

data = pd.read_csv('data.csv')
data2 = data[data['species_name'] == "Raflesia speciosa (Panay Rafflesia)"]
print(data2)

PANAY =[11.102947,122.516279]
Map=folium.Map(PANAY, zoom_start=9, tiles='Stamen Terrain')
Map

for i in range(0,len(data2)):

    folium.Marker([data2.iloc[i]["coordinates_lat"],data2.iloc[i]["coordinates_lng"]],tooltip=data2.iloc[i]["species_name"],popup='Date:''<br>'+str(data2.iloc[i]["date_report"])+'<br>''Sightings:''<br>'+str(data2.iloc[i]["no_observation"])).add_to(Map)
Map
```
