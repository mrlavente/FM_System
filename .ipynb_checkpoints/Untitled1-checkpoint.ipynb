```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
daf=pd.read_csv('deer.csv',parse_dates=['Date'])
df=daf[['Date','Population']]
df.info() # date format 
df.head() 
```

```python
df=df.groupby('Date').sum()
df=df.resample(rule='MS').sum() # rule & aggregation parameter
df.plot()
```

```python
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
seasonal_decompose(df,model='additive',period=4).plot();
```

```python
train=df[:30] #till Jul19
test=df[30:] # from aug19
```

```python
test
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
hwmodel=ExponentialSmoothing(train.Population,trend='add', seasonal='mul', seasonal_periods=4).fit()
```

```python
test_pred=hwmodel.forecast(12)
```

```python
test_pred
```

```python
train['Population'].plot(legend=True, label='Train', figsize=(10,6))
test['Population'].plot(legend=True, label='Test')

test_pred.plot(legend=True, label='predicted_test')
```
